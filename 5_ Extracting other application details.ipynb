```python
fields = ['appln_id', 'appln_auth', 'appln_filing_year', 'receiving_office', 'granted', 'docdb_family_size']
df2 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/tls201_part01.csv',skipinitialspace=True, chunksize=200000, header=0,usecols=fields)
header = True
chunkid=0
for chunk in df2:
    chunk = df1.merge(chunk, on='appln_id', how='inner') #filter by ids
    chunk.to_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/Adapt_details.csv', index=False, mode='a', header=header)
    print (chunkid)
    chunkid = chunkid + 1
    header = False
```

```python
del df1, df2, df3, df4
#merging with the larger file
df1 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/Adapt_details.csv',skipinitialspace=True, header=0)
if (df1['appln_id'].dtype != 'float64'):
            df1['appln_id'] = pd.to_numeric(df1['appln_id'], errors='coerce')
            df1.dropna(subset=['appln_id'], inplace=True)
            df1['appln_id'] = df1['appln_id'].astype('float64')
df2 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/Adapt_Residence_concise.csv',skipinitialspace=True, header=0)
df3 = pd.merge(df1, df2, on="appln_id", how = 'inner')
df3.to_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/Adapt_everything.csv', encoding='utf-8', index=False)
```

```python
import pandas as pd
import numpy as np
field = ['appln_id']
fields = ['appln_id', 'appln_auth', 'appln_filing_year', 'receiving_office', 'granted', 'docdb_family_size']
#add 'Build', 
Category = ['Sequester', 'ICT', 'Geo', 'Hydro', 'Ocean', 'Solar-T', 'PV', 'Thermal-PV', 'Wind', 'Combust', 'Fusion', 'Fission', 'Electric', 'Biofuel', 'Waste', 'Store- B', 'Store - C', 'Store- T', 'Store- M', 'Hydrogen', 'Other-1', 'Other-3', 'Processing', 'Transport', 'Wastewater', 'SmrtGrds', 'Other-4', 'Other-5', 'Nuclear', 'Non-Fossil', 'Enable','Other-6']
for c in Category:
    df1 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/' + c + '_Residence_concise.csv',skipinitialspace=True, usecols = field) 
    df2 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/tls201_part01.csv',skipinitialspace=True, chunksize=200000, header=0,usecols=fields)
    header = True
    chunkid=0
    for chunk in df2:
        chunk = df1.merge(chunk, on='appln_id', how='inner') #filter by ids
        chunk.to_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/' + c + '_details.csv', index=False, mode='a', header=header)
        print (str(chunkid) + c + str(1))
        chunkid = chunkid + 1
        header = False
    #similarly for the second third of the data
    df3 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/tls201_part02.csv',skipinitialspace=True, chunksize=200000, header=0,usecols=fields)
    header = True
    chunkid=0
    for chunk in df3:
        chunk = df1.merge(chunk, on='appln_id', how='inner') #filter by ids
        chunk.to_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/' + c + '_details.csv', index=False, mode='a', header=header)
        print (str(chunkid) + c + str(2))
        chunkid = chunkid + 1
        header = False
    #similarly for the last bit
    df4 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/tls201_part03.csv',skipinitialspace=True, chunksize=200000, header=0,usecols=fields)
    header = True
    chunkid=0
    for chunk in df4:
        chunk = df1.merge(chunk, on='appln_id', how='inner') #filter by ids
        chunk.to_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/' + c + '_details.csv', index=False, mode='a', header=header)
        print (str(chunkid) + c +str(3))
        chunkid = chunkid + 1
        header = False
    del df1, df2, df3, df4
    #merging with the larger file
    df1 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/' + c + '_details.csv',skipinitialspace=True, header=0)
    if (df1['appln_id'].dtype != 'float64'):
                df1['appln_id'] = pd.to_numeric(df1['appln_id'], errors='coerce')
                df1.dropna(subset=['appln_id'], inplace=True)
                df1['appln_id'] = df1['appln_id'].astype('float64')
    df2 = pd.read_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/' + c + '_Residence_concise.csv',skipinitialspace=True, header=0)
    df3 = pd.merge(df1, df2, on="appln_id", how = 'inner')
    df3.to_csv('/Volumes/NJ_4TB/PATSAT/MyFiles/' + c + '_everything.csv', encoding='utf-8', index=False)
    print(c + str(df1.shape[0]))
    print(c + str(df2.shape[0]))
    print(c + str(df3.shape[0]))
    del df1, df2, df3
```
